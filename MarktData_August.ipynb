```python
day_end_df["Timestamp"] = pd.to_datetime(day_end_df["Timestamp"],format= '%Y-%m-%d %H:%M:%S.%f')
day_end_df["Sales (net)"] = day_end_df["Sales (net)"].str.replace(',', '').astype(float)
#day_end_df["Sales (Number)"] = day_end_df["Sales (Number)"].str.replace(',', '').astype(int)
day_end_df["Sales (gross)"] = day_end_df["Sales (gross)"].str.replace(',', '.').astype(float)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
net_sales = [day_end_df["Timestamp"], day_end_df["Sales (net)"]]
net_sales
```

```python
ax = plt.gca()

#day_end_df.plot(kind='line',x='Timestamp',y='Sales (net)',ax=ax)
day_end_df.plot(kind='line',x='Timestamp',y='Sales (Number)',ax=ax)
day_end_df.plot(kind='line',x='Timestamp',y='Sales (gross)', color='red', ax=ax)

plt.show()
```

```python
article_turn_df = pd.read_csv('article_turn.csv', delimiter=";", encoding = 'ISO-8859-1')
article_turn_df.head(2)
```

```python
article_turn_df["Profit"] = article_turn_df["Profit"].str.replace(',', '.').astype(float)
article_turn_df.head()
```

```python
article_quantity = article_turn_df.groupby(['Article text, short'])['Quantity'].agg('sum').nlargest(10).reset_index()
article_quantity
```

```python
article_profit = article_turn_df.groupby(['Article text, short'])['Profit'].agg('sum').nlargest(10).reset_index()
article_profit
```

```python

article_profit[0]
```

```python
plt.figure(figsize=(15,8))

sns.barplot(x=article_profit['Article text, short'], y=article_profit.Profit[2:])
plt.xticks(rotation=70)
plt.tight_layout()
```

```python
plt.figure(figsize=(15,8))

sns.barplot(x=article_quantity['Article text, short'], y=article_quantity.Quantity)
plt.xticks(rotation=70)
plt.tight_layout()
```
